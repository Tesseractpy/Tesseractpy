```
df['clean_tweet'] = df['text'].apply(lambda x : ' '.join([tweet for tweet in str(x).split()if not tweet.startswith("@")]))
#Removing all the greek characters using unidecode library
df['clean_tweet'] = df['clean_tweet'].apply(lambda x : ' '.join([unidecode.unidecode(word) for word in str(x).split()]))
df['clean_tweet'] = df['clean_tweet'].apply(lambda x : ' '.join([word for word in x.split() if not word == 'h(m)+' ]))
d = {'luv':'love','wud':'would','lyk':'like','wateva':'whatever','ttyl':'talk to you later',
               'kul':'cool','fyn':'fine','omg':'oh my god!','fam':'family','bruh':'brother',
               'cud':'could','fud':'food','motherfucka':'motherfucker'}
df['clean_tweet'] = df['clean_tweet'].apply(lambda x : ' '.join(d[word] if word in d else word for word in x.split()))

r=df[df['label']==0]
r.head(50)
```

```
countneg=0
countpos=0

basepath='/content/test3/fake'
basepath1='/content/test3/real'
basepath2='/content/train3/fake'
basepath3='/content/train3/real'
for count in range(26030):
    if ((df['label'][count]==1) and (countneg%2!=0)):
        f = open(os.path.join(basepath,str(count)+".txt"), "a")
        f.write(df['clean_tweet'][count])
        f.close()
        print(basepath)
        countneg+=1
        continue

    if ((df['label'][count]==1) and (countneg%2==0)):
        f = open(os.path.join(basepath2,str(count)+".txt"), "a")
        f.write(df['clean_tweet'][count])
        f.close()
        print(basepath2)
        countneg+=1
        continue

    if ((df['label'][count]==0) and (countpos%2!=0)):
        f = open(os.path.join(basepath1,str(count)+".txt"), "a")
        f.write(df['clean_tweet'][count])
        f.close()
        print(basepath1)
        countpos+=1
        continue

    if ((df['label'][count]==0) and (countpos%2==0)):
        f = open(os.path.join(basepath3,str(count)+".txt"), "a")
        f.write(df['clean_tweet'][count])
        f.close()
        print(basepath3)
        countpos+=1
        continue

print('countpos ',countpos)
print('countneg', countneg)
```
